```python
import numpy as np

#import Pandas
import pandas as pd
```

## Import & Clean Data

```python
#read CSV files
entries = pd.read_csv("C:/Users/KaliO/OneDrive/Documents/Grad School/Fall 2022/DSCI 689/Bus Project/Bacis Analysis/entries.csv")
loops=pd.read_csv("C:/Users/KaliO/OneDrive/Documents/Grad School/Fall 2022/DSCI 689/Bus Project/Bacis Analysis/loops.csv")
```

```python
#merge csv files by loop_id
entries_loops = pd.merge(entries, 
                      loops, 
                      on ='loop_id', 
                      how ='inner')
entries_loops
```

### Rename Columns & Remove Unnecessary Columns

```python
entries_loops.rename(columns={'loop_name':'Loop Name', 'date_added':'Date', 'boarded':'Students Boarded'}, inplace=True)
```

```python
entries_loops.drop(['is_deleted_x', 'is_deleted_y'], inplace=True, axis=1)
```

### Group Data by Loop & Students Boarded

```python
#group & summarize by loop names
entries_loops.groupby('Loop Name')['Students Boarded'].sum()
```

### Group Data by Date & Loop Name

```python
by_date = entries_loops.groupby(['Date','Loop Name']).sum()
by_date
```

### Sort Students Boarded by Descending Order

```python
#sort ascending by date and descending by boarded
sby_date = by_date.sort_values('Students Boarded', ascending=False)
sby_date
```

### Include Month & Season Data

Convert timestamp into a date/time format

```python
entries_loops['timestamp']=pd.to_datetime(entries_loops['timestamp'])
entries_loops['Date']=pd.to_datetime(entries_loops['Date'])
```

Check Data Type of Field

```python
entries_loops.info()
```

Created New 'Month' Column based on timestamp

```python
entries_loops['Month']=entries_loops['timestamp'].dt.month_name()
entries_loops
```

```python
entries_loops['Season']=np.where((entries_loops['Date']>='3/20/2021') & (entries_loops['Date']<='6/19/2021'), 'Spring',
                                np.where((entries_loops['Date']>='6/20/2021') & (entries_loops['Date']<='9/21/2021'), 'Summer',
                                        np.where((entries_loops['Date']>='9/22/2021') & (entries_loops['Date']<='12/20/2021'), 'Fall',
                                                         'Winter')))
entries_loops
```
